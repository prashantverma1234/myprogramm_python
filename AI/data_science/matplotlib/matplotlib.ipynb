```python

def create_plot(ptype):
    #setting the x axis
    x=np.arange(-10,10,0.01)
    #setting the y axis
    
    if(ptype=='linear'):
        y=x
    elif(ptype=='quadratic'):
        y=x**2
    elif(ptype=='cubic'):
        y=x**3
    elif(ptype=='quartic'):
        y=x**4

    return(x,y)
    

# define figure of all plots
fig=plt1.figure()

# define subplots and their position in figure
plt2=fig.add_subplot(321) # 3 is a no. of rows and 2 is a no. of cols, 1 is a split sequece
plt3=fig.add_subplot(322)
plt4=fig.add_subplot(223)
plt5=fig.add_subplot(224)

#ploting point on each subplot
x,y=create_plot('linear')
plt2.plot(x,y,color='r')
plt2.set_title('$y_1=x$')

x,y=create_plot('quadratic')
plt3.plot(x,y,color='b')
plt3.set_title('$y_2=x^2$')

x,y=create_plot('cubic')
plt4.plot(x,y,color='g')
plt4.set_title('$y_3=x^3$')

x,y=create_plot('quartic')
plt5.plot(x,y,color='g')
plt5.set_title('$y_4=x^4$')

plt1.show()
```

```python

def create_plot(ptype):
    #setting the x axis
    x=np.arange(0,5,0.01)
    #setting the y axis
    
    if(ptype=='sin'):
        # a sin wave
        y=np.sin(2*np.pi*x)
    elif(ptype=='exp'):
        #negative exponetial function
        y=np.exp(-x)
    elif(ptype=='hybrid'):
        # a damped sin wavw
        y=np.sin(2*np.pi*x)*(np.exp(-x))
        

    return(x,y)
    



# define subplots and their position in figure
plt2=plt1.subplot2grid((11,1),(0,0),rowspan=3,colspan=1) 
plt3=plt1.subplot2grid((11,1),(4,0),rowspan=3,colspan=1)
plt4=plt1.subplot2grid((11,1),(8,0),rowspan=3,colspan=1)

#ploting point on each subplot
x,y=create_plot('sin')
plt2.plot(x,y,label="sine wave",color="r")

x,y=create_plot('exp')
plt3.plot(x,y,label="negative expontial",color='b')


x,y=create_plot('hybrid')
plt4.plot(x,y,label="damped sin wave",color='g')

plt2.legend()
plt3.legend()
plt4.legend()


plt1.show()
```
